выводим на экран названия столбцов с помощью df.columns

```python
df = df.rename(columns = {'trackname': 'Trackname', 'genre': 'Genre'})
```

```python
df
```

```python
df['Trackname']=df['Trackname'].fillna(0)
df['Artistname']=df['Artistname'].fillna(0)
df['Genre']=df['Genre'].fillna(0)
df['time']=df['time'].fillna(0)
```

```python
df.head(20)
```

проверка и устранения дубликатов

```python
df = df.drop_duplicates()
df = df.drop_duplicates().reset_index()
df
```

```python
df['time'] = pd.to_datetime(df['time'],errors = 'coerce')
df.info()
```

```python
data_pivot = pd.pivot_table(df, index=['Trackname'], values = ['time'])
sorted_df = data_pivot.sort_values(by=['time'])
sorted_df
```

```python
data_pivot = pd.pivot_table(df, index=['Trackname'], values = ['time'])
sorted_df = data_pivot.sort_values(by=['time'], ascending=False)
sorted_df
```

```python
n_by_state = df.groupby("time")["Trackname"].count()
n_by_state
```
